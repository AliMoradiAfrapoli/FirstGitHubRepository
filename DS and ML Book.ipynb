## Changing data type

```python
DICTgender = {1: 'Male', 2: 'Female'} # dictionary specifies replacement
DICTsituation = {1: 'Single', 2: 'With Spouse', 3:'With Family', 4:'With Else'} 
DICTmeat = {0:'Never', 1: 'Less Than Once a Week', 2: 'Once a Week', 3:'2-3 Times a Week', 4:'4-6 Times a Week', 5:'Every Day'} 
DICTfat = {1: 'Butter', 2: 'Margarine', 3:'Peanut Oil', 4:'Sunflower Oil', 5:'Olive Oil', 6:'Mix of Vegetable Oils (e.g., Isio4)', 7:'Colza Oil', 8:'Duck or Goose Fat'} 
```

```python
nutri['gender']=nutri['gender'].replace (DICTgender).astype('category')
nutri['situation']=nutri['situation'].replace (DICTsituation).astype('category')
nutri['meat']=nutri['meat'].replace (DICTmeat).astype('category')
nutri['fat']=nutri['fat'].replace (DICTfat).astype('category')
nutri['chocol']=nutri['chocol'].replace (DICTmeat).astype('category')
nutri['fish']=nutri['fish'].replace (DICTmeat).astype('category')
nutri['raw_fruit']=nutri['raw_fruit'].replace (DICTmeat).astype('category')
nutri['cooked_fruit_veg']=nutri['cooked_fruit_veg'].replace (DICTmeat).astype('category')
```

```python
nutri['weight']=nutri['weight'].astype('float')
nutri['height']=nutri['height'].astype('float')
nutri['age']=nutri['age'].astype('float')
```

```python
nutri.to_csv('nutri.csv', index=False)
```
